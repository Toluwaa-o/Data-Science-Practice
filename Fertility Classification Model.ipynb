```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
columns = ['Season', 'Age', 'Childish diseases', 'Accident or Trauma', 'Surgical Intervention', 'High fevers in last year', 'Frequency of alcohol consumption', 'Smoking habit', 'Number of hours spent sitting per day', 'Output']

df = pd.read_csv('fertility_Diagnosis.txt', names=columns)

df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
df['Output'] = enc.fit_transform(df['Output'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns=['Output'])
y = df['Output']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(x_train, y_train)
```

```python
log.score(x_train, y_train)
```

```python
log.coef_
```

```python
weight_table = pd.DataFrame(columns=['field', 'weight'])

weight_table['field'] = x.columns
weight_table['weight'] = log.coef_[0]
```

```python
weight_table
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = log.predict(x_test)

y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train, y_train)
```

```python
gnb.score(x_train, y_train)
```

```python
y_pred = gnb.predict(x_test)

y_pred
```

```python
accuracy_score(y_test, y_pred)
```
